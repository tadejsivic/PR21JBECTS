# KONČNO POROČILO

## PROBLEM
Problem predstavlja analizo registriranih vozil.

Ker je problem še kar obširen, smo se osredotočili na več podproblemov (Za narediti podrobnejšo analizo specifičnega problema, bi morali imeti drugačne podatke. Naši podatki so veliki, imajo dosti atributov, obširni).

Zato smo problem razdelili na manjše kose:
- Analiza starostnih skupin (Večji del seminarske je bilo raziskovnje značilnosti po starostnih skupinah)
- Analiza lastnosti vozil (Pogledali smo lastnosti, kot so poraba goriva, moč, starost,... vse to je sicer uporabljeno v povezavi s starostnimi skupinami)
- Analiza vpliva na okolje (Kako zeleni so naši avtomobili, v smislu CO2, CO ter trdnih delcev, ...)
- Analiza pravnih oseb (Nima svojega poglavja, kasneje izkazalo za neuporabno)

## Podatki
Podatki so bili v veliki zbirki (3 datotekah csv, vsaka po 650 000 vrstic, vsaka 108 stolpcev). Podatke smo izluščili iz ene datoteke, saj jih je bilo veliko. Če bi obravnavali vse podatke, bi porabili veliko več časa, razultati pa bi bili precej podobni (650 000 je reprezentativno za Slovenijo). Uporabili smo le 38 pomembnejših stolpcev, tako smo si količino podatkov še zmanjšali.

Kljub temu, da je zbirka uradna z Ministrstva za Infrastukturo, je še vedno ogromno napak. Poleg manjkajočih vrednosti (katere so razumljive), je veliko več človeških napak.

Ponekod so bile porabe goriva nenormalne, nekje veliko preveč sedežev kot je sploh možno, nekatere moči prevelike, starosti uporabnikov nemogoče (ali pa samo zgolj ilegalne). Predpriprava podatkov nam je torej vzela večino časa. Poleg priprave zbrike, je bilo ogromno filtriranja manjkajočih vrednosti ter nesmiselnih vrednosti.

- Young Adults [25-34]

### Moč in poraba vozil
Za vsako starostno skupino smo tudi analizirali, kako močna vozila uporabljajo. Tu so vzeta vsa vozila, torej električna vozila in tovornjaki skupaj z motornimi kolesi. Zanimalo nas je, ali Slovenci veliko gledamo na moč vozila med kupovanjem.

![](img\moc_starostne_skupine.png "Povprečna moč vozil za starostne skupine")

Črtkana črta predstavlja splošno povprečno moč vozila, torej namesto da gledamo točne številke, lahko primerjamo med seboj ter s povprečjem.
> Moč je v kilo-vatih [kW]

Starejši torej vozijo najšibkejša vozila.

Mladi tudi vozijo še kar šibka vozila. Ne smemo pozabiti, da tukaj gledamo vsa vozila, ne samo avtomobile. Tu so torej šibkejša vozila, saj je veliko med temi motornih koles.

S podobnim razmišljanjem preberemo kdo vozi najmočnejša vozila. Ljudje s starostne skupine (25-34) mladi odrasli vozijo najmočnešja vozila, ampak tu vmes so verjetno tudi tovornjaki. Močnejša vozila potrebujejo torej za delo ali pa so pač premožnejši od drugih skupin.

>Ali moč vpliva na odločitev pri nakupu? V bistvu kupi močnejši avto tisti, ki ga rabi. Torej nima vpliva, po navadi se človek v naprej izbere kaj potrebuje, ne pa obratno.

Povprečna poraba pa je bolj podobna med sabo

![](img\poraba_starostne_skupine.png "Povprečna poraba vozila za starostne skupine")

Ker se poraba ne razlikuje dosti, ne moremo izvleči kaj uporabnega.

Starejši ljudje imajo največjo porabo, kar je zanimivo, saj vemo iz prej, da vozijo najšibkejše avte.

Da ne delamo domnev brez kakršnekoli podlage, si poglejmo raje več analiz.


### Starost vozil
Če smo želeli narediti kakšne uporabne analize, smo morali tudi vedeti, koliko stara so vozila. Novejša vozila imajo drugačne mehanizme, so čistejša, varnejša, bolj ekonomična,...

Stolpični diagram za starosti:

![](img\povprecna_starost_avta.png "Povprečna starost vozil po starostnih skupinah")

Črtkana črta je splošna povprečna starost vozila, torej nekje 10.

Starejši ljudje vozijo tudi najstarejšja vozila. In to ne malo starejša, še kar starejša (oni verjetno tudi dvignejo splošno povprečje).

Odrasli imajo mlajša (podpovprečno stara vozila), kar je pričakovano, saj takrat človek zamenja največ vozil. 

Zelo mikavno je zdaj narediti kakšne hipoteze, ampak z več analizami bomo bolj uspešni.

### Prevožena razdalja
Poleg starosti vozila nam več pove dejanska prevožena razdalja. Sama starost ni toliko pomembna, če je avto bil v garaži pol leta. 
>Prevožena razdalja se meri v kilometrih [km]

Izračunali smo torej povprečno letno prevoženo razdaljo.

Če bi imeli podatke za prejšnja leta, bi lahko samo pogledali razliko prevožene razdalje letos in lani, a tukaj smo omejeni na to datoteko. Ni velik problem, letno prevoženo razdaljo lahko še vedno izračunamo, vsaj povprečno.

Povprečno letno prevoženo razdaljo dobimo tako, da celotno prevoženo razdaljo razdelimo na vsa leta (celotno starost vozila). To nam torej da število kilometrov/leto:

$\Large\overline{r}_{letna}=\frac{r_{celotna}}{s}$

kjer je  $r$: Razdalja, $s$: Starost vozila

Povprečna letna prevožena razdalja vozila, po starostnih skupinah:

![](img\prevozena_razdalja.png "Povprečna letna prevožena razdalja, po starostnih skupinah")

Mladi in mlajši odrasli imajo največ kilometrov letno, morda zaradi prevozov v službe, šole, ali pa samega užitka.

Starejši, medtem, pa imajo najmanjšo prevoženo razdaljo, morda gredo le redko v trgovino. Če povežemo prejšnja poglavja, lahko povzamemo, da starejši ljudje so zelo nepovezani s svojimi avti, saj jih imajo ogromno časa (največjo starost avtov), uporabljajo jih ne dosti (majhna razdalja), kljub temu pa imajo višjo porabo od vseh (način vožnje, srkrb za vozilo).

To je bila neke vrsta interpretacija, zaradi česa izvajamo podatkovno rudarjenje. Tako smo z grafi in povezovanjem le teh prišli do neke domneve.


### Registrske oznake
Na hitro lahko stopimo iz analiziranja lastnosti vozil na analizo registrskih tablic. Zakaj? To nam bo pomagalo razumeti, ali so vozila koncentrirana na nekem območju, ali uniformno porazdeljena.

Slika:

![](img\delez_registrske_oznake.png "Registrske oznake proti številu prebivalcev")

Slika vsebuje dva grafa, dva različna stolpca. 

Stolpec oranžne barve je št. prebivalcev v tisti regiji. To število smo dobili iz spleta.

Stolpec modre barve je število registrskih tablic za tisto regijo (torej KP, LJ, PO).

Drug ob drugem nam lepo prikažeta, v katerih regijah je več avtov kot ljudi in kje obratno.

V ljubljani je malo več avtov kot ljudi, ampak v Postojni jih pač ne registrirajo dosti. Pomembnejša špica je tudi v Kopru, kjer je dosti več vozil v primerjavi s prebivalci. 

Povprečen slovenec ima torej svoj avto, ali bolj znani pregovor, da povprečna družina ima 2 avta (povprečna družina je namreč sestavljena iz 2 staršev).

Zdaj ko vemo, da so vozila uniformno porazdeljena, lahko odmislimo vse osamelce.

### Izposojanje vozil
V podatkih o registraciji smo imeli dva podatka o človeku, ki je pripeljal vozilo na registracijo. Poleg fizične ali pravne osebe, poleg spola in starosti, smo imeli podatka o Lastniku in o Uporabniku vozila.
> Lastnik vozila ni nujno tudi uporabnik.
> En preprost primer je, ko študent uporablja avtomobil, ta pa je še vedno pisan na očeta. 

Pravilen izraz seveda ni *izposoja*, vendar moramo nekako povedati, da mislimo na to situacijo.

>Situacija, kadar uporabnik vozila ni hkrati tudi lastnik, se v nadaljevanju imenuje *izposoja*.

Katera starostna skupina bo torej imela največ izposojenih vozil?

![](img\izposoja.png "Delež ljudi, ki si izposoja vozila, po starostnih skupinah")

Ustavimo se malo. Diagram je narisan. In tudi ima smisla. Višina najvišjega stolpca pravi 20%. To ne pomeni, da ima vsak peti odrasel izposojen avto. To je delež vseh ljudi, ki si izposojajo. Drugače povedano, če imamo vse ljudi, ki izposojajo avte, je 20% možnosti da bo izbrani odrasel.

Diagram pa pravi, da odrasli (in mladi odrasli - po naši definiciji starostnih skupin) si največ izposojajo vozila. Pričakovali smo morda mlade. S prejšnjimi poglavji si lahko poskusimo pojasniti.

Mladih je pač dosti manj, odrasli pa tudi vozijo več in menjujejo dosti več vozil. Morda pa gre za stvar, ko sta v družini dva avta, oba pisana na moža, vendar enega uporablja žena. 

Razvidno iz diagrama je tudi, da so starejši ljudje res svoj razred. Avti so namreč večinoma njihovi, kar je tudi pojasnjeno s tem, da imajo najstarejše avte. 


### Uvoz
Vozila seveda uvažamo iz različnih držav. V sloveniji preprosto nimamo dosti tovarn za to. Zanimalo nas je, iz kje pa največ vozil pride?

![](img\uvozeni_drzava.png "Deležev uvoženih vozil iz posamezne države")

Verjetno nas je večina pričakovala Nemčijo, in to smo tudi dobili. Nemška tehnologija pač ne potrebuje opisa. Ampak kako bi to ugibali? Pri začetni analizi smo si pogledali, katere znamke vozil so najbolj priljubljene. Renault in Volkswagen sta daleč najbolj priljubljena. 

Volkswagen ima tovarno v Nemčiji (pa tudi Slovaški), Renault pa v Franciji.

Kar je zanimivo videti iz diagrama, je da nekaj vozil prihaja iz Jugoslavije. Zanimivo, ker pač te države ni več. Slovenci imamo pač ogromno traktorjev, in večina jih je ravno iz Jugoslavije. (Ali so proizvedeni novi zdaj ali pa še vedno tako stari traktorji; ni važno, teh podatkov ne bomo uporabljali)

### Barve
Preden gremo na bolj pomembno in podrobno analizo, pa lahko samo na hitro pogledamo...kakšne barve vozil imamo. Zdi se kot neuporabna analiza, vendar se je naš prvotni problem vrtel okoli vprašanja: "Kako se Slovenci odločimo kupiti vozilo?". Ker nimamo točnejših podatkov o nakupu avtov in o tem podobnem, lahko samo sklepamo.

Ali barva vozila vpliva na našo odločitev, da kupimo nek avto?

![](img\barve.png "Delež avtov z neko barvo")

Ne vem. Ali res vpliva?

Največ je sivih. Tukaj notri je seveda vpletena srebrna barva. Ampak vse barve se še ločijo na metalike, in kromirane in pastel in na še razne druge. Poleg tega ima npr. modra dosti več odtenkov. Siva in srebrna seveda niso toliko različne.

Tri najbolj popularne barve so torej siva, bela in črna. Niso kaj posebne barve. To je sicer nekaj, kar bi pričakovali. Imamo pa Slovenci dosti več "slabših" barv, torej zelenih in rjavih, v primerjavi svetovno (Hitro google iskanje pove, da je zelenih in rjavih avtov nekje pod 1%, pri nas sta obe barvi na nekje 5% vozil).

Kaj nam to pove? Ali barva vozila vpliva na nakup? Najverjetneje ne. Pri nakupu verjetno gledamo dosti več lastnosti vozila, kot so starost, prevoženi kilometri, poraba,... Ker pa je toliko vozil "slabših barv" (npr. zelene/rjave barve), ki jih je po svetu zelo malo, lahko povzamemo, da Slovenci gledamo dosti manj na barvo kot drugje po svetu. 

Če dobite za dobro ponudbo nek odličen avto, ga torej takoj vzemite, pa tudi če je zelen.&emsp;:)


### ONESNAŽEVANJE
Poleg vseh teh krasnih lastnosti vozil, pa imamo še en oddelek, ki je ostal za analizo. To morda ne zanima povprečnega človeka, ki kupi avto vsakih 5 let. Ampak nekoga, ki želi narediti spremembo, morda nekoga višje v politiki, pa to še kako zanima.

Kako zelena so vozila, ki jih uporabljamo v Sloveniji.

Tukaj smo se osredotočili predvsem na izpuste CO2 ter izpust trdih delcev pri avtomobilih z dizelskimi motorji.

##### CO2
Tukaj smo si pogledali, kako izpusti CO2 padajo skozi leta pri vseh novo registriranih vozilih. Pri tem smo pazili na nenormalne vrednosti ter omejili zgornjo mejo izpusta pri računanju letnega povprečja.

![](img\izpust_co2.png "Izpusti CO2 za novo registrirana vozila")

Razvidno je, da je skozi leta in z omejevanjem izpustov prišlo do izboljšav. 

Prav lepo je videti trendno črto kako potuje strmo navzdol. 

##### IZPUSTI CO2 PO AVTOMOBILSKIH ZNAMKAH
Izbrali smo nekaj popularnih avtomobilskih znamk ter primerjali njihove izpuste CO2. Pazili smo, da pri računanju povprečja izpusta CO2 skozi leta, posamezne znamke izberemo le osebne avtomobile, saj nekatere od teh proizvajajo tudi tovornjake, ter zgornjo mejo CO2 izpustov omejili na 300, ker gre tokrat le za osebne avtomobile.

![](img\izpust_co2_za_vozila.png "Izpusti CO2 za posamezna vozila")

Vidimo lahko, da imajo znamke kot so Renault, Toyota in Volkswagen nekoliko nižje izpuste kot Audi in Bmw, saj te dve znamki imata v svojih flotah večje in močnejše avtomobile, hkrati pa ponujata športne in še močnejše verzije posameznih modelov.

Te številke si je še vedno težje predstavljati, saj ne vemo kaj pomeni 150g CO2. Za lažjo predstavo si oglejmo Evropske številke.

![](img\izpustico2_evropaž.png "Izpusti CO2 - Evropa")

>Ta slika vzeta iz neta in ni del seminarske. Je malce starejša, črtkane črte pa najbrž napovedujejo vrednosti. Še vedno si je pa lepo ogledati, da primerjamo z našimi.

Nismo kaj dosti boljši. Naše črte so ponekod višje, ponekod podobne z evropskimi. Za onesnaževanje se seveda moramo truditi skupaj, iz tukaj le vidimo, da Slovenci kupujemo avte in ne gledamo toliko na izpuste. 

##### IZPUSTI TRDIH DELCEV NOVO REGISTRIRANIH VOZIL
Za prikaz smo izbrali le osebne avtomobile.

![](img\trdi_delci.png "Izpust trdih delcev")

Skozi leta je izpust zelo padel zaradi izboljšave filtrov, še posebej pa je lepo viden padec po uveljavi DPF filtra, ki je namenjen za zbiranje in izločanje prav teh trdih delcev.

Ta analiza ne velja ravno za slovenska vozila. V Sloveniji namreč ne proizvajamo svojih vozil. Je pa lepo narediti analizo na slovenskem trgu ter videti, da se vse lepo poveže z evropskim. 

Pomembno je, da nadaljujemo z razvojem ter tako pripomoremo k čistejšem in bolj zdravem okolju.



## Zaključek
Ta seminarska je bila na temi Registriranih vozil v Sloveniji (za letošnje leto).

Podatke smo na dolgo in široko obračali ter filtrirali. Večinoma zaradi tega, ker so bili podatki slabši (Kljub temu, da je 650k vrstic v samo eni datoteki, je razumljivo, da pride do napak. Ampak to izdaja Ministrstvo za Infrastrukturo, a ne bi poslali podatke skozi druge roke, da ne gre toliko napak skozi?).

Naredili smo nekaj manjših, zanimivejših analiz, katere so nam povedalo nekakšno zgodbico, kako živimo Slovenci.
Večina je bila branje povprečij iz grafa, ter interpretiranje z uporabo več grafov.

Sedaj veste, da naslednjič ko boste hodili ob cesti in se boste ozrli na njo, je velika možnost, da boste videli: 50-letnika v 10 let starem sivem Volkswagenu, ki se pelje iz Ljubljane.
